```python

def extract_features(file_path):
    y, sr = librosa.load(file_path, sr=None)  # Load audio file
    
    # Feature extraction
    features = []
    
    # 1. MFCC (Mel-Frequency Cepstral Coefficients)
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=10)
    mfcc_mean = np.mean(mfcc, axis=1)
    features.extend(mfcc_mean)
      
    return np.array(features)
```

```python
# Function to apply augmentation
def augment_audio(audio, sr):
    choice = random.choice(["pitch", "stretch", "noise", "shift", "volume"])
    
    if choice == "pitch":
        return librosa.effects.pitch_shift(audio, sr, n_steps=random.randint(-2, 2))
    elif choice == "stretch":
        return librosa.effects.time_stretch(audio, rate=random.uniform(0.8, 1.2))
    elif choice == "noise":
        noise = np.random.normal(0, 0.005, audio.shape)
        return audio + noise
    elif choice == "shift":
        shift_range = int(random.uniform(-0.2, 0.2) * sr)
        return np.roll(audio, shift_range)
    elif choice == "volume":
        return audio * random.uniform(0.7, 1.3)
    
    return audio  # Default return
```

```python
X_file = os.path.join(model_folder, "X.npy")
Y_file = os.path.join(model_folder, "Y.npy")

# Load or process audio files
if os.path.exists(X_file) and os.path.exists(Y_file):
    X = np.load(X_file, allow_pickle=True)
    Y = np.load(Y_file, allow_pickle=True)
    print("X and Y arrays loaded successfully.")
else:
    X = []  # input array
    Y = []  # output array
    for root, dirs, directory in os.walk(path):
        for j in range(len(directory)):
            name = os.path.basename(root)
            print(f'Loading category: {dirs}')
            print(name + " " + root + "/" + directory[j])
            if 'Thumbs.db' not in directory[j]:
                # Load audio file
                audio_path = os.path.join(root, directory[j])
                audio, sr = librosa.load(audio_path, sr=None)
                
                # Remove noise from the audio (assumed to be implemented)
                #audio = remove_noise(audio, sr)
                
                # Extract features
                mfccs = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=13)
                zero_crossings = librosa.feature.zero_crossing_rate(y=audio)
                spectral_centroid = librosa.feature.spectral_centroid(y=audio, sr=sr)
                chroma = librosa.feature.chroma_stft(y=audio, sr=sr)
                spectral_bandwidth = librosa.feature.spectral_bandwidth(y=audio, sr=sr)
                spectral_contrast = librosa.feature.spectral_contrast(y=audio, sr=sr)
                rms = librosa.feature.rms(y=audio)
                rolloff = librosa.feature.spectral_rolloff(y=audio, sr=sr, roll_percent=0.85)
                tonnetz = librosa.feature.tonnetz(y=audio, sr=sr)
                mel_spectrogram = librosa.feature.melspectrogram(y=audio, sr=sr)

                # Flatten and concatenate all features into a single feature vector
                features = np.concatenate((
                    mfccs.flatten(), 
                    zero_crossings.flatten(),
                    spectral_centroid.flatten(),
                    chroma.flatten(),
                    spectral_bandwidth.flatten(),
                    spectral_contrast.flatten(),
                    rms.flatten(),
                    rolloff.flatten(),
                    tonnetz.flatten(),
                    mel_spectrogram.flatten()
                ), axis=0)
                
                max_length = 3000  # Choose a consistent length for all feature vectors
                if len(features) < max_length:
                    features = np.pad(features, (0, max_length - len(features)), mode='constant')
                else:
                    features = features[:max_length]

                # Append the input audio features array to X
                X.append(features)
                # Append the index of the category in categories list to Y
                Y.append(categories.index(name))
    
    # Convert lists to numpy arrays
    X = np.array(X)
    Y = np.array(Y)
```

```python
algorithm_names = []
accuracy = []
precision = []
recall = []
fscore = []

labels = ['airleak', 'background noise', 'idling', 'normal', 'oil leak']


def performance_metrics(algorithm, y_pred, y_true):
    # Convert to NumPy int arrays (safe casting)
    y_true = np.array(y_true).astype(int)
    y_pred = np.array(y_pred).astype(int)

    # Metrics
    acc = accuracy_score(y_true, y_pred) * 100
    prec = precision_score(y_true, y_pred, average='macro', zero_division=0) * 100
    rec = recall_score(y_true, y_pred, average='macro', zero_division=0) * 100
    f1 = f1_score(y_true, y_pred, average='macro', zero_division=0) * 100

    # Save results
    algorithm_names.append(algorithm)
    accuracy.append(acc)
    precision.append(prec)
    recall.append(rec)
    fscore.append(f1)

    # Print
    print(f"\n{algorithm} Accuracy    : {acc:.2f}%")
    print(f"{algorithm} Precision   : {prec:.2f}%")
    print(f"{algorithm} Recall      : {rec:.2f}%")
    print(f"{algorithm} F1-Score     : {f1:.2f}%")

    print(classification_report(y_true, y_pred, target_names=labels, zero_division=0))

    # Confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=labels, yticklabels=labels)
    plt.title(f"{algorithm} Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.tight_layout()
    plt.show()
```

```python

AUDIO_FOLDER = "test data\\normal"
LABELS = ['Air leak', 'Background noise', 'idling', 'normal', 'Oil leak']

# Load RF model
rf_hubert = joblib.load("model/RF_on_HuBERT.pkl")


feature_extractor = Wav2Vec2FeatureExtractor.from_pretrained("facebook/hubert-base-ls960")
hubert = HubertModel.from_pretrained("facebook/hubert-base-ls960")
hubert.eval()

def hubert_embed(wav, sr, target_sr=16000):
    if sr != target_sr:
        wav = librosa.resample(wav, orig_sr=sr, target_sr=target_sr)
    inputs = feature_extractor(wav, sampling_rate=target_sr, return_tensors="pt")
    with torch.no_grad():
        out = hubert(inputs["input_values"]).last_hidden_state
        emb = out.mean(dim=1).squeeze().numpy()
    return emb

for filename in os.listdir(AUDIO_FOLDER):
    if not filename.endswith(".wav"):
        continue
    
    path = os.path.join(AUDIO_FOLDER, filename)
    y, sr = librosa.load(path, sr=None)

    emb = hubert_embed(y, sr).reshape(1, -1)
    pred = rf_hubert.predict(emb)[0]
    label = LABELS[pred]

    # Show waveform
    plt.figure(figsize=(10, 3))
    librosa.display.waveshow(y, sr=sr)
    plt.title(f"Predicted: {label}")
    plt.tight_layout()
    plt.show()

    print(filename, " → ", label)
```
